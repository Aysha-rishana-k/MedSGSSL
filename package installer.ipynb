```python
# ============================================
# üî• CUDA 11.8 ‚Äî Correct & Conflict-Free Install
# ============================================

INSTALL_ALL = True

if INSTALL_ALL:
    import sys, subprocess

    def pip_install(cmd):
        subprocess.check_call([sys.executable, "-m", "pip"] + cmd)

    print("üßπ Removing conflicting OpenCV packages (if any)...")
    for pkg in ["opencv-python", "opencv-python-headless"]:
        try:
            pip_install(["uninstall", "-y", pkg])
        except Exception:
            pass

    # -----------------------------
    # 1Ô∏è‚É£ Install PyTorch FIRST (CUDA 11.8)
    # -----------------------------
    print("üöÄ Installing PyTorch + torchvision with CUDA 11.8...")
    pip_install([
        "install", "-q",
        "torch", "torchvision", "torchaudio",
        "--index-url", "https://download.pytorch.org/whl/cu118"
    ])

    # -----------------------------
    # 2Ô∏è‚É£ Install remaining packages
    # -----------------------------
    print("üì¶ Installing remaining libraries...")
    pip_install([
        "install", "-q",
        "numpy", "pandas", "matplotlib", "seaborn",
        "scikit-learn", "opencv-python==4.7.0.72", "Pillow",
        "tqdm", "scipy", "kagglehub", "timm"
    ])

    print("‚úÖ CUDA 11.8 setup complete ‚Äî no conflicts!")

else:
    print("INSTALL_ALL is False ‚Äî skipping installations.")
```

```python
# ============================================
# ‚úÖ Step 1.2: Quick Dependency Check
# ============================================

packages = {
    "torch": "torch",
    "torchvision": "torchvision",
    "numpy": "numpy",
    "pandas": "pandas",
    "matplotlib": "matplotlib",
    "seaborn": "seaborn",
    "scikit-learn": "sklearn",
    "opencv": "cv2",
    "PIL": "PIL",
    "tqdm": "tqdm",
    "scipy": "scipy",
    "kagglehub": "kagglehub"
}

missing = []
for name, mod in packages.items():
    try:
        __import__(mod)
        print(f"‚úÖ {name} OK")
    except Exception as e:
        print(f"‚ùå {name} missing ({e.__class__.__name__})")
        missing.append(name)

if missing:
    print("\nSome packages are missing. You can set INSTALL_ALL = True in the install cell and re-run it, or install the listed packages manually.")
else:
    print("\nAll dependencies appear available.")
```
